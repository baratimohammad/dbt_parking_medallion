```python
sql_query = """
select
    violation_code,
    sum(fee_usd) as total_revenue_usd
from
    silver_parking_violation_codes
group by 
    violation_code
having 
    not(total_revenue_usd>=1)
"""

with duckdb.connect(database='data/nyc_parking_violations.db') as conn:
    display(conn.execute(sql_query).df())
```
